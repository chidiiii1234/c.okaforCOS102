```python
# Finding compound interest, simple interest and annuity plan
# Function to solve for simple interest 
def comp_int():
    p = float(input("Enter Principal: "))
    r = float(input("Enter Rate(in percentage): "))
    t = float(input("Enter Time(in years): "))
    ci = p * (1 + (r/100)*t)
    print(f"The Amount on Compound Interest is = {ci}") 

# Function to solve for simple interest
def simp_int():
    p = float(input("Enter Principal: "))
    r = float(input("Enter Rate(in percentage): "))
    t = float(input("Enter Time(in years): "))
    n = float(input("Enter the number of times the interest is compounded per year: "))
    si = p * ((1 + (r/n))**(n*t)) 
    print(f"The Amount on Simple Interest is = {si}")   

# Function to solve for annuity plan
def ann_plan():
    pmt = float(input("Enter payment per period: "))
    r = float(input("Enter Rate(in percentage): "))
    t = float(input("Enter Time(in years): "))
    n = float(input("Enter the number of payments per year: "))
    ap = pmt * (((1 + (r/100)/n)** (n*t) - 1 ) / ((r/100)/n))
    print(f"The Amount on Simple Interest is = {ap}")   
              
# Function to ask the user what calculations they would like to make
def get_input():
    choice = input("What would you like to calculate today(pls note you are only to use the codes provided) : ")
    return(choice)

# Function to ask whether they wish to make another calculation
def go_again():
    return(input("Would you like to make more calculations(Y/N): "))
    

print("Calculations      | Code ")
print("------------------|------")
print("Compound Interest | CI   ")
print("------------------|------")
print("Simple Interest   | SI   ")
print("------------------|------")
print("Annuity Plan      |  AP  ")

r2 = "Y"
while r2.upper() == "Y":
    r1 = get_input()

    if r1.upper() == "CI":
      comp_int()
                   
    elif r1.upper() == "SI":
      simp_int()

    elif r1.upper() == "AP":
      ann_plan()

    else:
       print("Error. wrong code") 
    r2 = go_again()
    print("Thank you and have a nice day")
```
