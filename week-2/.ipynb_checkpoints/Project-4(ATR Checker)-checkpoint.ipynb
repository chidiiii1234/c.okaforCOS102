```python
# Function to enable the user to input more than one staff's detail without re-running the code
def go_again():
    choice = input("Would you like to input the details of a different staff member(Y/N): ")
    return(choice)
e = "Y"
while e.upper() == "Y":
# Taking the age, years of experience and name of the staff members
    a = input(f"Hii, whats your first name: ")
    b = input(f"Hii, whats your last name: ")
    c = float(input(f"How old are you(in years): "))
    d = float(input(f"How many years of experience have you had: "))
# Computing the staff's ATR based on their details
    if c >= 55 and d > 25:
        print(f"{a} {b}, your Annual Tax Revenue(ATR) is 5.6 million naira")
    elif c >= 45 and 25 > d > 20:
        print(f"{a} {b}, your Annual Tax Revenue(ATR) is 4.48 million naira")
    elif c >= 35 and 20 > d > 10:
        print(f"{a} {b}, your Annual Tax Revenue(ATR) is 1.5 million naira")
    elif 35 > c and 10 > d:
        print(f"{a} {b}, your Annual Tax Revenue(ATR) is 5 hundred and 50 thousand naira")
    else:
        print("Sorry there's no ATR for you.")
    e = go_again()
    if e.upper() == "Y":
        print("Alrightyyy")
    else:
         print("Thank you for using our ATR checker")
```
