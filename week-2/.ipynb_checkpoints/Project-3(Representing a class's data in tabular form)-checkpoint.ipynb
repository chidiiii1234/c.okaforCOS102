```python
import pandas as pd
# Creating a dictionary containing the information of the girls in the class 
girls_data = {
    "Name": ["Evelyn", "Jessica", "Somto", "Edith", "Liza", "Madonna", "Waje", "Tola", "Aisha", "Latifa"],
    "Age": ["17", "16", "17", "18", "16","18", "17", "20", "19", "17"],
    "Height": ["5.5", "6.0", "5.4", "5.9", "5.6", "5.5", "6.1", "6.0", "5.7", "5.5"],
    "Scores": ["80", "85", "70","60", "76","66", "87", "95", "50", "49"]
}
# Loading the girls' dataframe from the Girls dictionary
df_girls = pd.DataFrame(girls_data, index =["001", "002", "003", "004", "005", "006", "007", "008", "009", "010"])
# Creating a dictionary containing the information of the boys in the class
boys_data = {
    "Name": ["Chinedu", "Liam", "Wale", "Gbenga", "Abiola", "Kola", "Kunle", "George", "Thomas", "Wesley"],
    "Age": ["19", "16", "18", "17", "20", "19", "16", "18", "17", "19"],
    "Height": ["5.7", "5.9", "5.8", "6.1", "5.9", "5.5", "6.1", "5.4", "5.8", "5.7"],
    "Scores": ["74", "87", "75", "68", "66", "78", "87", "98", "54", "60"]
}
# Loading the boys' dataframe from the Boys dictionary
df_boys = pd.DataFrame(boys_data, index =["011", "012", "013", "014", "015", "016", "017", "018", "019", "020"])

# Adding a new column for gender to the dataframes
df_girls["Gender"] = "Girl"
df_boys["Gender"] = "Boy"

# Concatinating the two dataframes to make one big dataframe that displays the whole class's data in tabular form 
df = pd.concat([df_girls, df_boys])

print(df)
```
