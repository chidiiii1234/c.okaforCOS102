```python
from sympy import symbols, solve, Rational
# Function to get the values of the coefficients
def get_input(label):
    user_input = input(f"Enter {label}: ")
    if '/' in user_input:
        return float(Rational(user_input))
    else:
        return float(user_input)

# Function to find the roots of a cubic equation
def cubic_roots(a, b, c, d):
    x = symbols("x")
    expr = a*(x**3) + b*(x**2) + c*x + d
    print(f"The roots of {a}*(x**3) + {b}*(x**2) + {c}*x + {d} are: {solve(expr)}")

# Functions to finds the roots of a quadratic equation
def quadratic_roots(a, b, c):
    x = symbols("x")
    expr = a*(x**2) + b*x + c
    print(f"The roots of {a}*(x**2) + {b}*x + {c} are: {solve(expr)}")

# Functions to find the roots of a quartic equation
def quartic_roots(a, b, c, d, e):
    x = symbols("x")
    expr = a*(x**4) + b*(x**3) + c*(x**2) + d*x + e
    print(f"The roots of {a}*(x**4) + {b}*(x**3) + {c}*(x**2) + {d}*x + {e} are: {solve(expr)}")

def go_again():
    choice = input("Would you like to calculate anything else?(Y/N): ")
    return(choice)

print("Hii what can we do for you today")
print("Calculation         | Code")
print("--------------------|-----")
print("Quardratic Equation |QDE  ")
print("Cubic Equation      |CE   ")
print("Quartic Equation    |QE   ")

r2 ="Y"
while r2.upper() == "Y":
    r1 = input("What would you like to solve today(enter the appropriate code):")
    if r1.upper() == "QDE":
        print("The standard form of a quadratic equation is ax^2 + bx + c")
        a = get_input("a")
        b = get_input("b")
        c = get_input("c")
        quadratic_roots(a, b, c)
        
    elif r1.upper() == "CE":   
        print("The standard form of cubic equation is ax^3 + bx^2 + cx + d")
        a = get_input("a")
        b = get_input("b")
        c = get_input("c")
        d = get_input("d")
        cubic_roots(a, b, c, d)

    elif r1.upper() == "QE":
        print("The standard form of quartic equation is ax^4 + bx^3 + cx^2 + dx + e") 
        a = get_input("a")
        b = get_input("b")
        c = get_input("c")
        d = get_input("d")
        e = get_input("e")
        quartic_roots(a, b, c, d, e)
    else: 
        print("Invalid code entered")
    r2 = go_again()    
    print("Thank for using our code!")
```
